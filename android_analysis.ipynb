```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Travel Industry Analysis/datasets/zxcbv.csv')
df.drop(columns=['index', 'lang'], inplace=True)
df.rename(columns={'AppRating': 'RatingScore'}, inplace=True)
```

```python
df['RatingScore'].fillna(0, inplace=True)
df.Downloads = df['Downloads'].str.replace(',+', '')
df['inAppPurchase'].fillna(False, inplace=True)
df['releaseDate'].fillna('', inplace=True)
```

```python
df.drop_duplicates(subset='AppID', inplace=True)
df.to_csv('/content/drive/MyDrive/Travel Industry Analysis/datasets/g_play_mined_data.csv')
```
