```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Travel Industry Analysis/datasets/apple_mined_data.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df['Ratings'].fillna(0, inplace=True)
df.Ratings = df.Ratings.replace({'K':'*1e3', 'M':'*1e6'}, regex=True).map(pd.eval).astype(int)
df['ratingScore'].fillna(0, inplace=True)
df['Downloads'].fillna(0, inplace=True)
df.Downloads = df.Downloads.replace({'K':'*1e3', 'M':'*1e6'}, regex=True).map(pd.eval).astype(int)
df['downloadSize'].fillna(0, inplace=True)
df.downloadSize = df['downloadSize'].str.strip(' MB')
df.rename({'downloadSize': 'downloadSize(MB)'}, inplace=True, axis=1)
df
```
